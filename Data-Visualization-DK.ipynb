```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plotter
%matplotlib inline  
# %matplotlib inline means all the plots will now show in notebook itself.
```

# Plotting using Functional Method

```python
# np.linspace(start, stop, num(default =50), others)
# Returns evenly spaced numbers over a specified interval.
x = np.linspace(0,10)
y = x**2
y = np.sqrt(x)
plotter.plot(x,y,'b')
plotter.xlabel("X Axis")
plotter.ylabel("Y Axis")
plotter.title("Functional Method Plot")
plotter.show()
```

# Plotting using Object Oriented Method

```python
figureOne = plotter.figure()
# figureOne.add_axes([left,bottom,width,height]
axes = figureOne.add_axes([0.2,0.1,0.7,0.9])
axes.plot(x,y)
axes.set_xlabel("X Axix")
axes.set_ylabel("Y Axis")
axes.set_title("Object Oriented Plot")
```
